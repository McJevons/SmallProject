```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 10000  # 本金
I = 0.05  # 年利率
i = I/12  # 月利率
M = 12  # 月数
```

每月还款额m_z=每月还款本金+每月还款利息。m_b+m_l

每月还款本金m_b=贷款总额/贷款月数 N/M。

每月还款利息m_l=贷款本金余额*贷款月利率 n*i（ 贷款月利率=年利率/12）。

贷款本金余额n=贷款总额-已还款月数 * 每月还款本金 N-m*N/M

```python
def bj(m):
    m_b = N/M
    n = N-m*N/M
    m_l = n*i
    m_z = m_b+m_l
    print('第%d月还款额：%f，还款本金：%f，还款利息，%f，期末余额：%f' % (m+1, m_z, m_b, m_l, n-m_b))
```

```python
for x in range(M):
    bj(x)
```

本月还息=本月余额×月利率

本月还本=每月还款额-本月还息

每月还款额=[贷款本金×月利率×（1+月利率）^还款月数]÷[（1+月利率）^还款月数－1]

```python
p=N*i*pow((1+i),M)/(pow((1+i),M)-1)
```

```python
def bx(M):
    l=None
    for m in range(M):
        if not l: # 期初余额
            m_s=N
        else:
            m_s=l[-1]
        m_l=m_s*i # 还息
        m_b=p-m_l # 还本
        m_e=m_s-m_b # 期末余额
        print('第%d月还款额：%f，还款本金：%f，还款利息，%f，期初余额：%f，期末余额：%f' % (m+1, p, m_b, m_l, m_s, m_e))
        l=[m+1, p, m_b, m_l, m_s, m_e]
        yield l
```

```python
bx_df=pd.DataFrame(columns=['月份','还款额','还本','还息','初期余额','末期余额'])
```

```python
bx_lst=[]
for x in bx(M):
    bx_lst.append(x)
```

```python
bx_df=pd.DataFrame(bx_lst,columns=['月份','还款额','还本','还息','期初余额','期末余额'])
```
