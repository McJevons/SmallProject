```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_male=pd.read_excel("2019上海国际马拉松成绩册.xls",sheet_name=0,header=None)
```

```python
data_female=pd.read_excel("2019上海国际马拉松成绩册.xls",sheet_name=1,header=None)
```

```python
data_male['性别']='男'
```

```python
data_female['性别']='女'
```

```python
data_male['性别名次']=data_male.index
```

```python
data_female['性别名次']=data_female.index
```

```python
data=pd.concat([data_male,data_female])
```

```python
data=data.sort_values([1]).reset_index(drop=True)
```

```python
data['总名次']=data.index
```
